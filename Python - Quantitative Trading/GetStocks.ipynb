## Import libaray

```python
import requests
from io import StringIO
import pandas as pd
import numpy as np
```

## Download data from Taiwan Stock Exchange Corporation

```python
SearchDate = '20181207'
DataFromWeb = requests.get('http://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + SearchDate + '&type=ALLBUT0999')
DataFromWeb.text[:100]
```

## Parse the data by Pandas

```python
temp_list = []

# 不含0050
for temp in DataFromWeb.text.split('\n'):
    if len(temp.split('",')) == 17 and temp[0] != '=':
        temp_list.append(temp)

target_dict = {ord(c): None for c in ' '}

String_Data = StringIO("\n".join([i.translate(target_dict) for i in temp_list]))

df = pd.read_csv(String_Data)

df[:5]
```

## Select Stocks

```python
# 用本益比<15的條件來篩選
df1 = df[pd.to_numeric(df['本益比'], errors='coerce') < 15]
df1[:10]
```

## Sort the Stocks

```python
df2 = df[pd.to_numeric(df['本益比'], errors='coerce') < 15].sort_values(by=['本益比', '收盤價'], ascending = False)
df2[:10]
```
